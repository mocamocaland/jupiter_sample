```python
# 中レベルのインターフェースを使用グラフ
from bokeh.plotting import output_notebook, figure, show

output_notebook()
p = figure(plot_width=200, plot_height=200)
p.line([1, 2], [1, 2])
p.circle([1.5], [1.5])
show(p)
```

```python
# 低レベル
from bokeh.plotting import output_notebook, figure, show
from bokeh.models import LinearAxis, Range1d

output_notebook()
p = figure(width=300, height=300, toolbar_location='above')
p.extra_y_ranges = {"small_data": Range1d(start=0, end=5)}
p.line([1, 2, 3],[1, 2, 3], y_range_name="small_data")
p.circle([1, 2, 3], [2000, 3000, 1000])
p.add_layout(LinearAxis(y_range_name="small_data"), 'right')
show(p)
```
