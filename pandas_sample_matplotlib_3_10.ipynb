```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ax = pd.Series([1, 2, 3]).plot()
ax.set_title('Line Chart')
plt.show()
```

```python
# スタイルの変更
plt.style.use('ggplot')
```

```python
# Seriesからplotメソッドを実行
ser = pd.Series([1, 2, 3])
ax = ser.plot()
ax.set_title('Line Chart')
plt.show()
```

```python
# DataFrameからplotメソッドを実行
df = pd.DataFrame({'a':[1, 2, 3], 'b':[3, 2, 1]})
ax = df.plot()
ax.set_title('Line Chart')
plt.show()
```
