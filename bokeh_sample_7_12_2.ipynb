```python
from bokeh.plotting import figure, output_notebook, show

output_notebook()
p = figure(plot_width=250, plot_height=250, toolbar_location='below')
show(p)
```

```python
# 指定したツールをツールバーに表示
p = figure(
    plot_width=250, plot_height=250,
    tools=['crosshair', 'wheel_zoom', 'reset'])
p.circle([1, 2, 3], [1, 2, 3], size=40)
show(p)
```

```python
# アクティブ状態を変更
p = figure(
    plot_width=400, plot_height=250,
    active_drag=None, active_scroll='wheel_zoom')
p.circle([1, 2, 3], [1, 2, 3], size=40)
show(p)
```
